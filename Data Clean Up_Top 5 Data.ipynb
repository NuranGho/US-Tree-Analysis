```python
import pandas as pd
import os
import us
```

```python
combined_data = pd.read_csv('combined_data.csv')
```

```python
drop_columns = combined_data.columns[12:]

# Drop the columns
combined_data = combined_data.drop(columns=drop_columns)
```

```python
combined_data.head()
```

```python
combined_data.info()
```

```python
# Group by the specified columns and count the occurrences of tree_name
grouped_df = combined_data.groupby(['state', 'common_name','scientific_name']).size().reset_index(name='tree_count')

# rename common_name
grouped_df = grouped_df.rename(columns={'common_name': 'tree_name'})
grouped_df = grouped_df.rename(columns={'scientific_name': 'tree_species'})
```

```python
grouped_df['state_code'] = grouped_df['state'].apply(lambda x: us.states.lookup(x).abbr if us.states.lookup(x) else x)
```

```python
grouped_df.to_excel('grouped_data.xlsx',index=False)
```

# Top 5 trees per state data

```python
# Group by 'state' and get the top 5 'tree_count' values per state
top_5df = grouped_df.groupby('state', group_keys=False).apply(lambda x: x.nlargest(5, 'tree_count'))
```

```python
top_5df.to_excel('top5_data.xlsx',index=False)
```
