```python
import pandas as pd
import os
import altair as alt
```

```python
current_directory = os.getcwd()

# List CSV files in the current directory
csv_files = [file for file in os.listdir(current_directory) if file.endswith('.csv')]

combined_data = pd.DataFrame()

for file in csv_files:
    data = pd.read_csv(file)
    combined_data = combined_data.append(data, ignore_index=True)
```

```python
combined_data.head()
```

```python
drop_columns = combined_data.columns[12:]

# Drop the columns
combined_data = combined_data.drop(columns=drop_columns)
```

```python
combined_data.head()
```

```python
df2 = combined_data.fillna("null")
```

```python
df2.info()
```

```python
df2 = df2.drop_duplicates()
```

```python
df2.to_csv('combined_data.csv', index=False)
```
