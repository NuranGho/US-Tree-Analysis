```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("combined_data.csv")
df
```

```python
df = df.fillna("null")
```

```python
tree_counts_per_state = df.groupby(['common_name', 'state']).size().reset_index(name='count')
tree_counts_per_state = tree_counts_per_state.sort_values(by=['count'], ascending=False)
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 'null']
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 0]
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 443]
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 444]
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 445]
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 446]
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= 448]
tree_counts_per_state = tree_counts_per_state[tree_counts_per_state['common_name']!= "*calls"]
tree_counts_per_state = tree_counts_per_state[~tree_counts_per_state['common_name'].str.contains("vacant", case=False)]
tree_counts_per_state = tree_counts_per_state.head(20)
tree_counts_per_state
```

```python
brush = alt.selection_interval()

points = alt.Chart(tree_counts_per_state).mark_point().encode(
    x=alt.X('state', axis=alt.Axis(title='State')),
    y=alt.Y('count:Q', axis=alt.Axis(title="Count")),
    color=alt.Color('common_name:N', legend=alt.Legend(title="Tree Species"))
).add_params(
    brush
)

bars = alt.Chart(tree_counts_per_state).mark_bar().encode(
    x=alt.X('count:Q', axis=alt.Axis(title="Count")),
    y=alt.Y('common_name:N', axis=alt.Axis(title="Species Name")),
    color='common_name:N'
).transform_filter(
    brush
)

composite_chart = (points & bars).properties(
    title=alt.TitleParams(
        text="Tree Species Count by State", 
        fontSize=18,  
        anchor='middle'  
    )
)

composite_chart
```
